Una caminata aleatoria 1D es un proceso estocástico donde una partícula se mueve sobre una línea, dando pasos de tamaño fijo hacia la izquierda o derecha con ciertas probabilidades en cada unidad de tiempo.


---



**Ejercicio:** Elaborar un código que implemente una camita aleatoria 1D, con probabilidad p de moverse a la derecha y probabilidad q de moverse a la izquierda, tal que p + q = 1.

```python
#importamos las paqueterías necesarias a utilizar
import random as rd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def caminata(n, p):
  pos=0
  camino=[pos]
  for _ in range(n):
    if rd.random()<p:
      pos+=1
    else:
      pos-=1
    camino.append(pos)
  return camino
```

```python
n= 500  # Número de pasos
p=0.5 # probabilidad p de moverse a la derecha

# Ejecutar la caminata aleatoria usando la función 'caminata'
cam = caminata(n,p)

# Visualizar la caminata
plt.figure(figsize=(10, 5))
plt.plot(cam, color='purple')
plt.title(f'Caminata Aleatoria 1D')
plt.xlabel('Número de Pasos')
plt.ylabel('Posición')
plt.grid(True)
plt.show()
```

¿Qué debe pasar con p para regresar al origen?

```python
n= 500  # Número de pasos
p=0.5 # probabilidad p de moverse a la derecha

# Ejecutar la caminata aleatoria usando la función 'caminata'
cam3 = caminata(n,p)

# Visualizar la caminata
plt.figure(figsize=(10, 5))
plt.plot(cam3, color='purple')
plt.title(f'Caminata Aleatoria 1D')
plt.xlabel('Número de Pasos')
plt.ylabel('Posición')
plt.grid(True)
plt.show()
```

¿Cuál es la posicion promedio de la caminata?


```python
pos_f=[]

for _ in range(1000):
  cams = caminata(1000,0.5)
  pos_f.append(cams[-1])
  plt.plot(cams)
plt.show()

media=np.mean(pos_f)

print('media de la posición final de 10 caminatas es', media)
```

Dando varioas valores para la cantidad de simulaciones que se hacen en el código, pude notar que cuando mayor era el número de las simulacoines, la media del último valor es cercana a 0, y cuando eran menos este valor variaba más, aunque claro, dando valores mayores nada asegura que la media sea cero.

¿Cuál es valor máximo y mínimo que puede tomar?
